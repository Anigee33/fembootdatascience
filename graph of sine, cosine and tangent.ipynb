```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(-20,20)
```

```python
y=np.sin(x)
z=np.cos(x)
a=np.tan(x)
```

```python
plt.plot(x,y,label='Sin of x')
plt.title('Graph of sin x')
plt.legend();
```

```python
plt.plot(x,z,label='cos of x')
plt.title('Graph of cos x')
plt.legend();
```

```python
plt.plot(x,a,label='tan of x')
plt.title('Graph of tan x')
plt.legend()
plt.grid(True);
```
